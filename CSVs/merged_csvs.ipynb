```python
import pandas as pd
```

```python
def cleaned_data(csv_name):
    df = pd.read_csv(csv_name)
    df.dropna(inplace=True)
    df = df.sample(5000)
    return df
```

```python
jan_22 = cleaned_data("202201-citibike-tripdata.csv")
feb_22 = cleaned_data("202202-citibike-tripdata.csv")
mar_22 = cleaned_data("202203-citibike-tripdata.csv")
apr_22 = cleaned_data("202204-citibike-tripdata.csv")
may_22 = cleaned_data("202205-citibike-tripdata.csv")
jun_22 = cleaned_data("202206-citbike-tripdata.csv")
jul_22 = cleaned_data("202207-citbike-tripdata.csv")
aug_22 = cleaned_data("202208-citibike-tripdata.csv")
sep_22 = cleaned_data("202209-citibike-tripdata.csv")
oct_22 = cleaned_data("202210-citibike-tripdata.csv")
nov_22 = cleaned_data("202211-citibike-tripdata.csv")
dec_22 = cleaned_data("202212-citibike-tripdata.csv")
```

```python
list_22 = [jan_22,feb_22,mar_22, apr_22, may_22, jun_22, jul_22, aug_22, sep_22, oct_22, nov_22, dec_22]
```

```python
citibike_2022 = pd.concat(list_22, ignore_index=True)
```
